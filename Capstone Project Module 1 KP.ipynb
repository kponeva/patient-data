```python
#Create a sample data on a dict list of 'data_patient'
data_patient = [   
    {
        "Name": "Hypatia",
        "Age": 28,
        "Gender": "Female",
        "Blood type": "0"
    },
    {
        "Name": "Susan",
        "Age": 29,
        "Gender": "Female",
        "Blood type": "A"
    },
    {
        "Name": "Bob",
        "Age": 30,
        "Gender": "Male",
        "Blood type": "AB"
    },
    {
        "Name": "Stiv",
        "Age": 31,
        "Gender": "Male",
        "Blood type": "AB"
    }
]

#Create the 'read_data' function to show patient's data.
def read_data(): 
    menu_read =('''
    ============= Patient Data Record =============
    1. Show all data
    2. Show selected data
    3. Back to main menu
    ''')
    option_read = int(input(menu_read + "Choose the menu items above [1-3]: ")) 
    #nested conditional statement to define each menu options
    if option_read in [1,2]: 
        if option_read == 1: #n input equal to what's on the list
            data = data_patient 
        else: #statement for selected data option
            data = [] 
            n = input("\nEnter the patient's name you're looking for: ") 
            for i in data_patient: 
                if n == i["Name"].lower(): #if n input equal to the value under 'name' key
                    data.append(i) 
        if len(data) == 0: #n input not equal to what's on the list
            print("\nThe name you input is invalid. Try again. ")
        else: #n input equal to what's on the list
            a = 0
            print(" ")
            print("\nPatient data \n\n")
            for x in data:
                a += 1
                menu_read = """{}. Name: {}; Age: {}; Gender: {}; Blood type: {} """.format(a, x["Name"], x["Age"], x["Gender"], x["Blood type"])
                print(menu_read)
    elif option_read == 3:
        return

    return read_data()
            
    
#Create the 'add_patient' function to add a new data.
def add_patient():
    #use while loop for continuous call
    while True:
        menu_add = '''
        ============= Adding new patient data =============
        1. Add a new patient
        2. Back to main menu
        '''
        
        print(menu_add)
        
        option_add = int(input("Please choose the menu items above [1-2]: "))
        #conditional statement to define each menu options
        if option_add == 1:
            name = None
            age = None
            gender = None
            blood = None
            agree = None
            
            while True:#use while true for adding new data + nested conditional statement
                if name == None: #if statement for name input
                    option_add = input("\nInsert the new patient's name here: ")
                    t = True 
                    for data in data_patient:
                        if data["Name"] == option_add:
                            print("{} is already in our database".format(option_add))
                            t= False
                            break
                    if t == False:
                        continue
                    name = option_add
                if age == None: #if statement for age input
                    option_add = int(input("Insert patient's age: "))
                    age = option_add
                if gender == None: #if statement for gender input
                    option_add = input("Patient's gender: ")
                    if option_add.lower() not in ["male", "female"]:
                        print("Choose either male or female")
                        continue
                    gender = option_add
                if blood == None: #if statement for blood input
                    option_add = input("Patient's blood type: ")
                    if option_add not in ["AB", "A", "B", "O"]:
                        print("Choose A, B, AB, or O")
                        continue
                    blood = option_add
                if agree == None: #if statement for confirmation
                    confirm = input("\nAre you sure the details given above are correct? (yes/no)")
                    if confirm == "yes":
                        new_data = {"Name": name, "Age": age, "Gender": gender, "Blood type": blood}
                        data_patient.append(new_data)
                        print("The new data is now successfully added.")
                    elif confirm == "no":
                        print("The new data input is cancelled.")
                    elif confirm not in ["yes", "no"]:
                        continue
                    break
        elif option_add == 2:
            break
        else:
            continue

    return
            
#Create the 'update_patient' function to make changes the existing data. 
def update_patient():
    while True: #use while loop + nested if statement to update data on the list
        menu_update = ('''
        ============= Updating patient data =============
        1. Update existing patient data
        2. Back to main menu
        
        ''')
        print(menu_update)
       
        option_update = int(input("Please choose the menu items above [1-2]: "))
        #conditional statement for each menu options
        if option_update == 1:
            num = None
            name = None
            age = None
            gender = None
            blood = None
            agree = None
            print(" ")
            for n, x in enumerate(data_patient): #allows us to loop over data patient and keep track of how many iterations have occurred
                menu_read = """{}. Name: {}; Age: {}; Gender: {}; Blood type: {} """.format(n+1, x["Name"], x["Age"], x["Gender"], x["Blood type"])
                print(menu_read)
                print(" ")

            while True: #while loop + nested if statement to indicate each items of data_patient list
                if num == None:
                    num = int(input("Choose the index number of the patient (max {}): ".format(len(data_patient))))
                    if num > len(data_patient):
                        print("It has only {} patient(s) on the list".format(len(data_patient)))
                        num = None
                        continue
                data_x = data_patient[num -1]
                print(data_x)
                if name == None:
                    x = input("Update the patient's name/skip: ")
                    if x.lower() == 'skip':
                        name = data_x["Name"]
                    else:
                        name = x
                if age == None:
                    x = int(input("Update the patient's age: "))
                    age = x
                if gender == None:
                    while True:
                        x = input("Define the gender (Male or Female) or skip: ")
                        if x.lower() == 'skip':
                            gender = data_x['Gender']
                            break
                        elif x.lower() in ['male', 'female']:
                            gender = x
                            break
                        else:
                            print("Input should be male/female or skip")
                            continue
                    
                if blood == None:
                    while True:
                        x = input("Patient's blood type or skip: ")
                        if x.lower() == 'skip':
                            blood = data_x["Blood type"]
                            break
                        elif x in ["AB", "A", "B", "O"]:
                            blood = x
                            break
                        else:
                            print("Input should be AB, A, B, O or skip")
                            continue

                if agree == None:
                    x = input("Are you sure to update the data? (yes/no)")
                    if x.lower() == 'yes': # to show the updated data before the user confirm
                        data_x = {"Name": name, "Age": age, "Gender": gender, "Blood type": blood}
                        data_patient[num -1] = data_x
                        print("Data update succeeded!")
                    elif x.lower() == 'no':
                        print("Data update cancelled.")
                    else:
                        continue
                    break
                
        elif option_update == 2:
            break
        else:
            continue
            

#Create the 'delete_data' function to show patient's data.
def delete_patient():
    #while loop to continues statement for deleting data menu
    while True:
        menu_delete = '''
        ============= Deleting patient data =============
        1. Delete existing patient data
        2. Back to main menu
        
        '''
        print(menu_delete)
        
        option_del = int(input("Please choose the menu items above [1-2]: "))
        if option_del == 1:
            print(" ")
            for num, x in enumerate(data_patient): #to show data patient before
                menu_read = """{}. Name: {}; Age: {}; Gender: {}; Blood type: {} """.format(num+1, x["Name"], x["Age"], x["Gender"], x["Blood type"])
                print(menu_read)
            while True:
                x = int(input("Choose the index number of the patient you want to delete: "))
                if x > len(data_patient):
                    print('It has only {} patient(s) on the list'.format(len(data_patient)))
                    continue
                break

            while True:
                confirm = input("Are you sure to delete this data? (yes/no): ")
                if confirm.lower() == 'yes':
                    del data_patient[x-1]
                    print("Data deleted. Thank you.")
                elif confirm.lower() == 'no':
                    print("Deleting data cancelled.")
                else:
                    continue
                break
                        
        if option_del == 2:
            break
        else:
            continue
                        
#Create the 'main_menu' function to show the menu of patient's data.
                        
def main_menu():
    main_option = '''
    ============= Alexandria Hospital Patient Record =============
        1. Show patients data
        2. Add new patients data
        3. Update patients data
        4. Delete patients data
        5. Exit
    
    '''
    print(main_option)
    menu = int(input("Choose the main menu item [1-5]: "))
    if menu == 1:
        read_data()
    elif menu == 2:
        add_patient()
    elif menu == 3:
        update_patient()
    elif menu == 4:
        delete_patient()
    elif menu == 5:
        print("Thank you and goodbye!")
        return 
    else :
        print("Wrong choice")
    
    return main_menu()

main_menu()
                        
```
